```python
dados2013 = pd.read_csv('DOM2013.csv',sep=',')
novo_dados2013 = dados2013.loc[:,["UF","V0226","V02270","V02322","V4614","V4621"]]

#Mudando a categoria das variáveis
novo_dados2013.V0226 = novo_dados2013.V0226.astype("category")
#novo_dados2013.V02270 = novo_dados2013.V02270.astype("category")
novo_dados2013.V02322 = novo_dados2013.V02322.astype("category")
#novo_dados2013.V4614 = novo_dados2013.V4614.astype("category")
#novo_dados2013.V4621 = novo_dados2013.V4621.astype("category")


#Rotulando os valores
novo_dados2013.V0226.cat.categories = (['sim', 'não'])
#novo_dados2013.V02270.cat.categories = (['sim', 'não'])
novo_dados2013.V02322.cat.categories = (['sim', 'não'])
#novo_dados2013.V4614.cat.categories = (['sim', 'não'])
#novo_dados2013.V4621.cat.categories = (['sim', 'não'])

#Rotulando as colunas
novo_dados2013 = novo_dados2013.rename(columns={"V0226":"Televisão_em_cores"})
novo_dados2013 = novo_dados2013.rename(columns={"V02270":"Número_de_televisões_de_tela_fina"})
novo_dados2013 = novo_dados2013.rename(columns={"V02322":"Tem_acesso_a_Internet"})
novo_dados2013 = novo_dados2013.rename(columns={"V4614":"Rendimento_Domiciliar"})
novo_dados2013 = novo_dados2013.rename(columns={"V4621":"Rendimendo_per_capita"})


novo_dados2013=novo_dados2013[novo_dados2013["Rendimento_Domiciliar"]!= 999999999999.0]
novo_dados2013=novo_dados2013[novo_dados2013["Rendimendo_per_capita"]!= 999999999999.0]

novo_dados2013.to_csv("resumo.csv")

novo_dados2013
```
